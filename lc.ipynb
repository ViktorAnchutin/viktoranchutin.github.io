```python
class Solution(object):
    def solve(self, i, nums, mem):
        if i == len(nums): return 1
        if mem[i] == 0:
            max_l = 0
            for k in range(i+1,len(nums)):
                if nums[k] > nums[i]:
                    max_l = max(self.solve(k,nums,mem), max_l)
            mem[i] = max_l + 1
        return mem[i]

    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        mem = [0]*len(nums)
        return max(self.solve(i,nums,mem) for i in range(nums))
        
        

            

Solution().lengthOfLIS([10,9,2,5,3,7,101,18])
```
